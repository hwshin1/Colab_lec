```python
loss
```

```python
x1 = np.random.randn(len(x))
```

```python
fig, loss_ax = plt.subplots(figsize=(8, 6))
acc_ax = loss_ax.twinx()
loss_ax.plot(x, loss(x), 'y', label='train loss')
loss_ax.plot(x, loss(x-x1/5), 'r', label='val loss')
acc_ax.plot(x, acc(x), 'b', label='train acc')
acc_ax.plot(x, acc(x-x1/7), 'g', label='val acc')

for label in acc_ax.get_yticklabels():
    label.set_color('blue')
loss_ax.set_xlabel('epoch', size=15)
loss_ax.set_ylabel('loss', size=15)
acc_ax.set_ylabel('accuracy', size=15, color='blue')
loss_ax.legend(loc='upper right')
acc_ax.legend(loc='lower right')
plt.show()
```

시계열 데이터 = 추세 + 계절성 + 잔차


```python
from statsmodels.tsa.seasonal import STL
```

```python
decomposition = STL(df['Passengers'], period=12).fit()
```

```python
decomposition
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, sharex=True, figsize=(10, 8))
ax1.plot(decomposition.observed, markevery=np.arange(6, 145, 12), marker='o')
ax1.set_ylabel('Observed', size=10)

ax2.plot(decomposition.trend)
ax2.set_ylabel('trend', size=10)

ax3.plot(decomposition.seasonal)
ax3.set_ylabel('seasonal', size=10)

for i in np.arange(0, 145, 12):
    ax3.axvline(i, linestyle='--', color='red', linewidth=1)

ax4.plot(decomposition.resid)
ax4.set_ylabel('Residuals', size=10)

plt.xticks(np.arange(0, 145, 12), np.arange(1949, 1962, 1)) # 1949년부터 1962년까지
fig.autofmt_xdate() # fmt를 자동으로 써라 xdata에 대해서
plt.tight_layout()
plt.show()
```

 # 모델로 훈련할 이전 데이터는 어떻게 샘플링 하는건가요?
 => 과거데이터 다 사용

```python
from scipy.stats import norm
import matplotlib.gridspec as gridspec
from scipy import stats

sns.set_style("whitegrid")
def visualize_target(y):
    fig = plt.figure(constrained_layout=True, figsize=(12,6))
    grid = gridspec.GridSpec(ncols=5, nrows=5, figure=fig)
    ax1 = fig.add_subplot(grid[0:2, :4])
    ax1.set_title('Histogram')
    sns.distplot(y, fit=norm, ax=ax1, color="indianred") # 나중에 가면 distplot이라는 메서드가 없어짐 사용하면 경고 메세지 뜸
    ''' 대체 코드
    sns.histplot(y, kde=True, stat='density', ax=ax1, color='indianred')

    xmin, xmax = ax1.get_xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, np.mean(y), np.std(y))
    ax1.plot(x, p)
    '''
    ax2 = fig.add_subplot(grid[2:, :4])
    stats.probplot(y, plot = ax2)
    ax2.set_title('QQ_plot')

    ax3 = fig.add_subplot(grid[:, 4])
    ax3.set_title('Box Plot')
    sns.boxplot(y=y, orient='v', ax = ax3,color='indianred')
```

```python
y = df['Passengers']
```

```python
visualize_target(y)
```
