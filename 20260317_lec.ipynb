# 파일 입출력
- 파일을 열어서 읽고, 쓰고, 덧붙이는 방법
    - `open(filename, mode)` 내장 함수로 file 이름을 가진 file 객체를 얻어 낸다.
    - 얻어진 파일 객체에서 자료를 읽거나, 쓰거나, 덧붙이는 작업 수행
    - 모든 작업이 끝나면 close() 하여, 작업 프로세스의 자원 점유 해제


|Mode|설명|
|---|---|
|r|읽기전용|
|w|쓰기전용|
|a|파일 끝에 추가|
|rb|이진 파일 읽기|
|wb|이진 파일 쓰기|
|ab|이진 파일 끝에 추가|

````python
import time
def outer(function):
    def inner(*args, **kwargs):
        start_time = time.time()
        function(*args, **kwargs) # 장식되는 함수에서 인수를 사용가능하게 함
        end_time = time.time()
        evaluation_time = (end_time - start_time) * 100
        return evaluation_time
    return inner
````

````python
# extand 하는 함수, copy 하는 함수
sample_list = list(range(10000))

@outer
def extend_test(sample_list):
    extend_list = []
    extend_list.extend(sample_list)

@outer
def copy_test(sample_list):
    copy_list = sample_list.copy()
````

````python
extend_test(sample_list) / copy_test(sample_list)
````

````python
ratio = extend_test(sample_list) / copy_test(sample_list)

if ratio > 1:
    print(f"extend가 copy보다 느림")
else:
    print(f"extend가 copy보다 빠름")
````

````python
# decorators
def wrapper(func):
    def wrappered_func():
        print(" == before ==")
        func()
        print(" == after ==")
    return wrappered_func

def myfunc():
    print('I am here.')

result = wrapper(myfunc)
print(result())
````

````python
# @ at이라고 읽자 @붙으면 이름이 붙은 함수에 들어간다. @ : 장식자
@wrapper
def myfunc2():
    print('me too')
myfunc2()
````
