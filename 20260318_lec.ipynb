- 클래스는 새로운 이름 공간을 지원하는 단위
- 이 이름 공간에는 함수와 변수가 포함될 수 있다.
- 클래스를 정의하는 것은 새로운 자료형을 하나 만드는 것이다.
- 인스턴스는 이 자료형의 객체롤 생성하는 것이다.
- 클래스 인스턴스 : 클래스를 호출하여 생성된 객체이다.

- 클래스 인스턴스 객체 : 클래스 인스턴스와 같은 의미이다. 인스턴스 객체라고 부르기도 한다.
- 멤버 : 클래스 혹은 클래스 인스턴스 공간에 정의된 변수이다.
- 메서드 : 클래스 공간에 정의된 함수 이다.
- 속성 : 맴버와 메서드 전체를 가리킨다.

````python
result = 0
def adds(num):
    '''
    계산기 모델링 한 함수
    입력
        NUM = 숫자
    출력
        result = 합계합
    예제
        adds(3)
        3
    ''' # API 정의
    global result
    result += num
    return result
````

클래스 이름 : FourCal

입력 자료 : 정수 2개

methods : 덧셈, 뺄쎔, 곱셈, 나눗셈

````python
class FourCal:
    def sums(self, n1, n2):
        self.value = n1
        return self.value + n2
    def minus(self, n1, n2):
        self.value = n1
        return self.value - n2
    def multi(self, n1, n2):
        self.value = n1
        return self.value * n2
    def div(self, n1, n2):
        self.value = n1
        if n2 == 0:
            return ('0으로 나누기 할 수 없습니다.')
        return self.value / n2
````

클래스 이름 : StopWatch

````python
# 내꺼
import time
class StopWatch(list):
    def __init__(self, v):
        self.value = v
    def extend_check(self):
        start = time.time()
        self.extend(self.value)
        end = time.time()
        return end - start
    def copy_check(self):
        start = time.time()
        self.value.copy()
        end = time.time()
        return end - start
    def extend_copy_check(self):
        result = self.extend_check() / self.copy_check()
        if result > 1:
            return ("extend가 copy보다 느림")
        return ("extend가 copy보다 빠름")
````

````python
# 강사님 버전
import time

class StopWatch:
    '''
    시간만 측정하기
    '''
    def __init__(self):
        '''
        시작타임
        '''
        self._createTime = time.time()
    def elapsedTime(self):
        '''
        현재타임에서 시작타임 빼기
        '''
        return time.time() - self._createTime
    def intervalTime(self, func, lists):
        func(lists)
        return self.elapsedTime()
````
