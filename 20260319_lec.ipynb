# 탐색적 데이터 분석의 시작은 그래프 부터

matplotlib는 아래와 같이 크게 두가지 방법으로 사용할 수 있습니다.
- stateless API(objected-based)
- stateful API(state-based)

stateless 방법은 내가 지정한 figure, 내가 지정한 ax에 그림을 그리는 방법입니다.

stateful 방법은 현재의 figure, 현재의 ax에 그림을 그리는 방법입니다.
