# Pandas marge 연습

```python
left = pd.DataFrame({'key': ['K0', 'K4', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on='key')
```

```python
pd.merge(left, right, how='left', on='key') # 교집합 왼쪽
```

```python
pd.merge(left, right, how='right', on='key') # 교집합 오른쪽
```

```python
pd.merge(left, right, how='outer', on='key') # 합집합
```
