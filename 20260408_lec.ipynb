```python
df = titanic.loc[:,['age', 'fare']][:10]
```

```python
df
```

```python
def add_10(n):
    return n + 10
```

```python
def add_two_obj(a, b):
    return a - b
```

```python
df_10 = df.age.apply(add_10)
```

```python
df_10
```

```python
df_10.apply(add_two_obj, b=10)
```

```python
df.age.apply(lambda n : n + 10)
```

```python
df.age.apply(lambda a, b : a - b, b = 10)
```

```python
def over_thirty(age):
    return age > 30
```

```python
df.age.apply(over_thirty)
```

```python
df.age.map(over_thirty)
```

map : 시리즈에 적용.

apply : 시리즈, 데이터프레임 모두 적용.

applymap : 개별값에 적용할 때 주로 사용.

```python
df.age.map(lambda age : True if age > 30 else False)
```

```python
def calculate_stats(col):
    max_val = col.max()
    min_val = col.min()
    mean_val = col.mean()
    return pd.Series([max_val, min_val, mean_val], index=['maxs', 'mins', 'means'])
```

```python
result_df = df.apply(calculate_stats, axis=0)
```
