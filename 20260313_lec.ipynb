1. 함수 이름을 철자 틀리지 않고, 대소문자 구분해서 그대로 적기
2. 괄호() 작성

## 함수란?
- 프로그램에서 어떤 일을 처리하기 위해 관련된 문을 모아 놓은 단위
- 함수를 사용하여 큰 프로그램을 작은 단위로 나눠 코드할 수 있다.
- 함수를 사용하여 프로그램을 논리적으로 구성할 수 있다.
- 함수를 사용하여 프로그램의 유지 보수를 쉽게 할 수 있다.
- 함수는 단순 코드의 반복을 줄이고, 재사용 가능하다.

## 코드를 논리적으로 분리 가능하면 --> 함수화

arguments : 함수를 호출 하는 입장에서, 데이터를 함수에 전달하기 위해 사용하는 변수나 값

함수 호출 : 정의되어 있는 함수를 사용하기 위해서는 함수를 호출해야 한다. 이때, 함수 호출이란, 함수 이름을 쓰고, 그 함수에 정의되어 있는 parameter에 맞게 argument를 적어주는 일을 말한다.

## return 문
- return 문으로 바노한되는 값들은 튜플로 구성하여 전달 된다.
- return 문이 없을 경우에는 None이 반환된다. (None = 아무 값이 없음을 표현하는 객체)
- 인수가 없는 경우의 return 문에서도 None이 반환된다.

# 함수의 인수
- 인수 : 함수를 호출할 때 전달되는 변수
- 위치 인수 : 인수의 위치로 값을 전달하는 방식
- 인수의 기본값 : 인수에 자신의 기본값을 갖고 있는 것
- 키워드 인수 : 인수 이름으로 값을 전달하는 방식
- 일반적으로 함수를 호출할 때 키워드 인수의 위치는 위치 인수 이후이다.
- 키워드 인수 이후에 순서에 의해서 인수 일치를 시킬 수는 없다.
- 가변 인수 리스트
    - 고정되지 않은 수의 인수를 함수에 전달하는 방법
    - 인수목록에 반드시 넘겨야 하는 고정 인수를 우선 나열하고, 나머지를 튜플 형식으로 한꺼번에 받는다.
    - 가변인수 부분은 *args 의 형식으로 지정한다.
- 정의되지 않은 키워드 인수
    - 키워드 인수를 이용해서 함수를 호출할 때, 미리 정의되어 있지 않은 키워드 인수를 받으려면, 함수를 정의할 때 마지막에 **kwargs 형식으로 기술한다.
    - 전달 받는 형식 : 사전
    - key => 변수명으로, value => 값으로 전달 된다.
- 인수의 순서 : 꼭 필요한 인수, \*args, \**kwargs

- argument 를 " = " 없이 사용하면 positional argument.
- "="을 사용해서 값을 정했으면 keyword argument.
- parameter 를 " = " 없이 사용하면 positional parameter.
- "="을 사용해서 값을 정했으면 keyword parameter.

# 문제
농장에 소, 닭이 있다

소 + 닭 머리수 = 20
소 + 닭 다리수 = 56
이 때 소가 몇마리 닭 몇마리

````python
for s in range(1, 21):
    for d in range(1, 21):
        if s * 4 + d * 2 == 56:
            if s + d == 20:
                print(s, d)
````

````python
def brute_force_algorithm(sum_head, sum_leg):
    for s in range(1, sum_head + 1):
        for d in range(1, sum_head + 1):
            if (s * 4) + (d * 2) == sum_leg:
                if s + d == sum_head:
                    return s, d

result = brute_force_algorithm(20, 56)
s = result[0]
d = result[1]
print(f"동물 마리수 (소 {s}마리, 닭 {d}마리)")
````

````python
def brute_force_algorithm(head, leg):
    for i in range(0, head + 1):
        j = head - i
        s = head - i - j
        condition = (i * 4) + (j * 2)
        if condition == leg:
            return i, j
    return(None, None)
````

````python
def barnyard():
    head = int(input('Enter the number of heads: '))
    leg = int(input('Enter the number of legs: '))
    (chickens, cows) = brute_force_algorithm(head, leg)
    if chickens != None:
        print('Number of chickens : ', chickens)
        print('Number of cows : ', cows)
    else:
        print('There is no solution')
````

````python
barnyard()
````

````python
def brute_force_algorithm(sum_head, sum_leg):
    for c in range(sum_head + 1):
        for d in range(sum_head + 1):
            s = sum_head - c - d
            if (c * 4) + (d * 2) + (s * 8) == sum_leg:
                if c + d + s == sum_head:
                    return f"동물 마리수 (소 {c}마리, 닭 {d}마리, 거미{s}마리)"
````

````python
result = brute_force_algorithm(8, 48)
print(result)
````

````python
# 값 2개
def brute_force_algorithm(sum_head, sum_leg):
    result = []

    for c in range(sum_head + 1):
        for d in range(sum_head + 1):
            s = sum_head - c - d

            if s < 0:
                continue

            if (c * 4) + (d * 2) + (s * 8) == sum_leg:
                result.append((c, d, s))

    return result
````

# 유효한 비밀번호 검증

1. 4글자 이상
2. 영문자 포함
3. 숫자 포함
4. 특수문자 !, @, #, $만 가능 --> in 연산자
5. 2~4번에 해당하는 문자들만 포함

포인트 => 검사한 횟수하고 문자 갯수하고 맞춰야 한다?

````python
def is_valid_password(password):
    if len(password) < 4:
        return False
    has_alpha = False
    has_digit = False
    allowed_special = "!@#$"
    for ch in password:
        if '0' <= ch <= '9':
            has_digit = True
        elif ('a' <= ch <= 'z') or ('A' <= ch <= 'Z'):
            has_alpha = True
        elif ch in allowed_special:
            pass
        else:
            return False
    if has_alpha and has_digit:
        return True
    else:
        return False
k=is_valid_password(input("비밀번호를 입력하세요: "))
print(k)
````

````python
def word():
    count = 0
    in_password = input('비밀번호를 입력하세요. : ')
    if len(in_password) < 4:
        return ('비밀번호는 4글자 이상이어야 합니다.')

    if in_password.isalpha():
        count += 1
        return True
    if in_password.isnumeric():
        count += 1
        return True
    if in_password in ['!', '@', "#", "$"]:
        count += 1
        return True

    if count == 3:
        return ('사용 가능한 비밀번호 입니다.')
````
