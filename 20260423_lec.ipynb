# 논리게이트

```python
# 이 AND함수는 돌려보면 참 거짓 참 거짓을 나타내는 것과 같다
# 논리 게이트를 방정식으로 만든셈
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.dot(w, x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
AND(0, 0), AND(0, 1), AND(1, 0), AND(1, 1)
```

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.dot(w, x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
NAND(0, 0), NAND(0, 1), NAND(1, 0), NAND(1, 1)
```

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.dot(w, x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
OR(0, 0), OR(0, 1), OR(1, 0), OR(1, 1)
```

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

```python
XOR(0, 0), XOR(0, 1), XOR(1, 0), XOR(1, 1)
```

# 고유값

## 고유값 분해
정방 행렬 A를 다음과 같이 분해

## $$
A = Q \boldsymbol{\Lambda} Q^{-1} \\
\boldsymbol{\Lambda} \ : \ 고유값을\ 대각선에\ 배치한\ 대각\ 행렬
$$

Q: A의 고유 백터를 열방향으로 배치한 행렬

Q-1 : Q의 역행렬

## 특잇값 분해
정방 행렬이 아닌 m*n 일반 행렬 A에 대해

## $$
A = U 𝚺 V^{T}
$$

U : A A.T의 고유벡터를 열에 배치한 m*m 행렬

V : A.T A의 고유벡터를 열에 배치한 n*n 행렬

Sigma : A A.T의 고유값의 제곱근을 대각선에 배치한 m*n 대각행렬

```python
def AND(x1, x2):
    w1 = 0.5
    w2 = 0.5
    x = np.array([x1, x2])
    w = np.array([w1, w2])
    b = -0.7
    tmp = np.dot(w, x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
from matplotlib import rc
import matplotlib.font_manager as fm
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def sigmoid_diff(x):
    return sigmoid(x) * (1 - sigmoid(x))
def sigmoid_tangent(x):
    return sigmoid_diff(x), sigmoid(x) - sigmoid_diff(x) * x
```

```python
x = np.arange(-6.0, 6.0, 0.1)
y1 = sigmoid(x)
a2, b2 = sigmoid_tangent(0)
y2 = a2 * x + b2
a3 = (sigmoid(2.5) - sigmoid(0)) / 2.5
y3 = a3 * x + b2
plt.plot(x, y1, label='y=f(x)')
plt.plot(x, y2, color='black', label='analytical')
plt.plot(x, y3, color='green', label='approximation')
xv = np.arange(-0.1, 0.5, 0.01)
plt.text(-0.2,0,"x")
plt.plot(np.zeros_like(xv), xv, 'k--')
xhv = np.arange(-0.1, sigmoid(2.5), 0.01)
plt.text(2,0,"x+h")
plt.plot(np.ones_like(xhv)*2.5, xhv, 'k--')
plt.scatter([0],[b2],color='red')
plt.ylim(-0.1,1.1)
plt.xlim(-4,4)
plt.legend(loc='upper center')
plt.show()
```
