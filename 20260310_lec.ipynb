```python
import numpy as np

def AND(x1, x2): # : 의미 => 새로운 suit(옷 한벌)가 시작됨을 알리는 지시자
# 수트 맞춰 주세요 들여쓰기 맞춰주세요 같은 의미
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.7
  tmp = np.dot(x,w) + b
  if tmp < 0:
    return 0
  else :
    return 1
```

```python
print(AND(0, 0))
print(AND(1, 0))
print(AND(0, 1))
print(AND(1, 1))
```
