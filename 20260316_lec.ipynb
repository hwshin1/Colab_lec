# Python Closure
- 함수 클로저란 함수가 참조할 수 있는 심볼 테이블 혹은 참조 환경이 함수와 함께 제공 되는 것
- 함수 클로저 조건
    1. 함수가 중첩
    2. 내부 함수에서 외부 함수의 값을 참조
    3. 외부 함수가 내부 함수를 반환해줘야 한다.

# Decorators(데코레이터)
```
A decorator is a closure function that takes a function(or a callable object) as an argument and return a function(or a callable object).

데코레이터는
함수(또는 호출 가능한 객체)를 인자로 받아서
함수(또는 호출 가능한 객체)를 반환하는 클로저 함수다.
```

기존 함수를 바꾸지 않고 기능을 추가할 수 있게 만드는 elapsed 함수와 같은 클로저를 데코레이터(decorator)라고 한다.

<<점프투 파이썬>>
```python
# decorator.py
import time

def elapsed(original_func):   # 기존 함수를 인수로 받는다.
    def wrapper():
        start = time.time()
        result = original_func()    # 기존 함수를 수행한다.
        end = time.time()
        print("함수 수행시간: %f 초" % (end - start))  # 기존 함수의 수행시간을 출력한다.
        return result  # 기존 함수의 수행 결과를 반환한다.
    return wrapper

def myfunc():
    print("함수가 실행됩니다.")

decorated_myfunc = elapsed(myfunc)
decorated_myfunc()
```

# 도박 시뮬레이션

자본금 = 10 money

목표 = 20 goal

할수있는 베팅 횟수 = 1000 trials

**== 중간과정 ==**

실행할 때마다 잃는 돈  = 1

lose_cash 변수 : 중간에 돈 계산

bets : 도박 머신을 돌린 횟수

wins : 승리 한 횟수

**==출력 값==**

- 이길 확률 : wins // trials * 100

- 평균 베팅수 : bets // trials

- 이기고 지는거 표현

`random.randrange(0,2) => 지면 1 이기면 0`

````python
import random
````

````python
def simulation():
    myMoney = 10
    goals = 20
    trials = 1000
    wins = 0
    bets = 0
    for i in range(trials):
        if random.randrange(0,2) == 1:
            myMoney = myMoney - 1
        if random.randrange(0,2) == 0:
            wins += 1
            myMoney = myMoney + 1
        if myMoney == goals:
            wins += 1
            break
````

````python
simulation()
````

````python
# 강사님 버전
stake = 5
goals = 25
trials = 1000
wins = 0
bets = 0

for t in range(trials):
    cash = stake
    while cash > 0 and cash < goals:
        bets += 1
        if random.randrange(0,2) == 0:
            cash += 1
        else:
            cash -= 1
    if cash == goals:
        wins += 1

print("이길 확률 : ", (100*wins//trials), '% wins')
print(bets//trials, '평균 베팅')
````

````python
# decorators
def wrapper(func):
    def wrappered_func():
        print(" == before ==")
        func()
        print(" == after ==")
    return wrappered_func

def myfunc():
    print('I am here.')
````

````python
result = wrapper(myfunc)
````

````python
print(result())
````
