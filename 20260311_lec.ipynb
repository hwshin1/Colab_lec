```python
def AND(x1, x2):
  '''
  AND : and 연산을 논리게이트

  입력 자료
  x1 : 정수 혹은 실수
  x2 : 정수 혹은 실수

  출력 자료
  0 혹은 1
  '''
  x = numpy.array([x1, x2])
  w = numpy.array([0.5, 0.5])
  b = -0.7
  tmp = numpy.dot(x, w) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```
